Write a program that will do the following:

1. Take a number input from the user.
2. Compute the factorial of that number.
3. You may not use any libraries for this program.  
4. Print the factorial for the number input to the user.

### Caution:  There is an upper limit to the number that can be computed.

N Factorial = N! = N(N-1)(N-2)...(1)

```python
"""
Write a program that will do the following:

    Take a number input from the user.
    Compute the factorial of that number.
    You may not use any libraries for this program.
    Print the factorial for the number input to the user.

"""

def main():
    
    num = 0
    num = float(input('Enter the number (factorial Exercise): ')) #instruction 1
    factorial = compute_factorial(num) # instruction 2
    print (f'{num} has a factorial of {factorial}') #instruction 4
    
def compute_factorial(n): #instruction 3 (write own function)
    if n == 1:
        return n
    elif n < 1:
        return ("NA")
    else:
        return n*compute_factorial(n-1)

main()

```
